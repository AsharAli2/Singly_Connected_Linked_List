# 5. *insert_before( )*

```python
class node():
    def __init__(self, data):
        self.data = data # attribute to store the value of the current node
        self.next = None # attribute to store the reference to the next node in the list

class LinkList():
    def __init__(self): # Initializes an empty linked list with the head attribute  None
        self.head = None # initialize an empty linked list with the head attribute  None

    def add(self, data):  # add and insert_at_end is same
    #Create a new node with the given data and adds it to the end of the list.
        new_node = node(data)
        if self.head is None:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def display(self):
        # Prints the data of each node in the linked list, starting from the head.
        new = self.head
        while new:
            print(new.data)
            new = new.next

    def search(self, data):
        # Search for the given data in the linked list, and prints "found" if the data is found in any of the nodes.
        new = self.head
        while new:
            if new.data == data:
                print("found")
            new = new.next
# ---------------------------------------------------
    def insert_before(self, data, x):
            # Creates a new node with the given data, and places it before the node with value x in the list.
            if self.head is None:
                print("Linked List is empty!")
                return
            if self.head.data == x:
                new_node = node(data)
                new_node.next = self.head
                self.head = new_node
                return
            n = self.head
            while n.next is not None:
                if n.next.data == x:
                    break
                n = n.next
            if n.next is None:
                print("Node is not found!")
            else:
                new_node = node(data)
                new_node.next = n.next
                n.next = new_node
# ---------------------------------------------------
s = LinkList()
s.add(10)  # add and insert_at_end is same
s.add(20)
s.add(30)
s.add(40)
s.add(50)
print("All the added elements:")
s.display()
print("\nSearch(30):")
s.search(30)
print("\nInsert 15 before 20:")
s.insert_before(15, 20)  # first argument is new element and second argument is where to put new element before
s.display()
```
